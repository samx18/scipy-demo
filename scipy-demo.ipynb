# Scipy Demo

## Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

## Statistic Distributions

```python
# generate values 
x = np.linspace(-6,6, 1000)
```

```python
x.shape
```

```python
# calculate PDF
fx = norm.pdf(x, loc=0, scale=1)
```

```python
fx.shape
```

```python
# If you pass one argument, it goes on the y axis
plt.plot(fx);
```

```python
# if you use two values, the first argument goes on the x axis and the second on y
plt.plot(x,fx);
```

```python
# calculate CDF
Fx = norm.cdf(x, loc=0, scale=1)
```

```python
plt.plot(x,Fx);
```

```python
#log pdf
logfx = norm.logpdf(x, loc=0, scale=1)
```

```python
plt.plot(x,logfx);
```

## Convolutions

```python
from PIL import Image
```

```python
# read the image
im = Image.open('R2D2-test.png')
```

```python
im.shape
```

```python
plt.imshow(im);
```

```python
# Convert to grayscale
gray = np.mean(im, axis=2)
```

```python
gray.shape
```

```python
plt.imshow(gray);
```

```python
plt.imshow(gray, cmap='gray');
```

```python
# create a gaussian filter using scipy
x= np.linspace(-6,6,50)
fx = norm.pdf(x,loc=0,scale=1)
flit = np.outer(fx,fx)
```

```python
# plut the filter as an image
plt.imshow(flit, cmap='gray')
```

```python
# import the  2d convolutional filter
from scipy.signal import convolve2d
```

```python
# apply to the image and filter
out = convolve2d(gray,flit)
```

```python
plt.subplot(1,2,1)
plt.imshow(gray)
plt.subplot(1,2,2)
plt.imshow(out);
```
